```python
import pandas as pd
review = pd.read_json('Downloads/Data/review_city.json',lines=True)
business = pd.read_json('Downloads/Data/business_city.json', lines=True)
tip = pd.read_json('Downloads/Data/tip_city.json',lines=True)
user = pd.read_json('Downloads/Data/user_city.json', lines=True)
business.head()
```

```python
# merge data
df_raw = pd.merge(review, business, how='left', on='business_id')
df_raw.head()
```

```python
# drop useless columns
from collections import Counter
df = df_raw.drop(columns=['business_id', 'review_id', 'user_id', 'funny', 'cool', 'date', 'latitude', 'longitude', 'is_open', 'review_count', 'hours'])
cate = Counter([j for k in [i.split(", ") for i in map(str,df["categories"])] for j in k]).most_common()
```

```python
# select one supecific restaurants "Burger" and remove duplicates lines
burger_index = [i.find("Burgers")!=-1 for i in map(str,df["categories"])]
burger = df[burger_index]
burger.drop_duplicates(subset='text', keep='first', inplace=True)
burger.shape
```

```python
import csv
burger_noreview = burger.iloc[:,0:10].drop(columns="text")
burger_noreview.to_csv("burger_noreview.csv", index=False)
```

```python
#remove punctuation
#lower 
text = list(burger.text)
import string
text = ["".join([char for char in a if char not in string.punctuation]) for a in text]
text = [i.lower() for i in text]
text[0]
```

```python
#remove non-English words
import nltk 
Eng_dir = set(nltk.corpus.words.words())
text_Eng = [" ".join([w for w in nltk.word_tokenize(j) if w in Eng_dir or not w.isalpha()]) for j in text]
```

```python
# replace those with "no", "not" to "no_xx"
text_no = [i.replace(" no "," no_").replace(" not "," no_").replace(" dont "," no_").replace(" isnt "," no_") for i in text_Eng] 
```

```python
# tokenize
words = [nltk.word_tokenize(i) for i in text_no]
words[0:3]
```

```python
# rwmove stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
filter_words = [[j for j in i if j not in stop_words] for i in words]
```

```python
#Lemmatize
#nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
lmtzr = WordNetLemmatizer()
lemmatized_words = [[lmtzr.lemmatize(x) for x in a] for a in filter_words]
with open("lemmatized_words.txt","w", encoding='utf-8') as f:
    w = csv.writer(f,delimiter=',',lineterminator='\n')
    w.writerows(lemmatized_words)
```

```python
# count frequency
from collections import Counter
import json
lemmatized_words_unlist = [i for item in lemmatized_words for i in item] 
counter_words = Counter(lemmatized_words_unlist).most_common()
with open("counter_words.txt","w", encoding='utf-8') as f:
    w = csv.writer(f,delimiter=',',lineterminator='\n')
    for key, value in counter_words:
        w.writerow([key, value])
```
